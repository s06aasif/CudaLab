# Tried to give some colors to Segmented Images 

```python
from PIL import Image, ImageEnhance, ImagePalette

def enhance_segmentation_image(image_path):
    palette = [
        (0, 0, 0),        # Unlabeled
        (70, 70, 70),     # Building
        (100, 40, 40),    # Fence
        (55, 90, 80),     # others
        (220, 20, 60),    # Pedestrian
        (153, 153, 153),  # Pole
        (157, 234, 50),   # RoadLine
        (128, 64, 128),   # Road
        (244, 35, 232),   # SideWalk
        (107, 142, 35),   # Vegetation
        (0, 0, 142),      # Vehicles
        (102, 102, 156),  # Wall
        (220, 220, 0),    # TrafficSign
        (70, 130, 180),   # Sky
        (81, 0, 81),      # Ground
        (150, 100, 100),  # Bridge
        (230, 150, 140),  # RailTrack
        (180, 165, 180),  # GuardRail
        (250, 170, 30),   # TrafficLight
        (110, 190, 160),  # Static
        (170, 120, 50),   # Dynamic
        (45, 60, 150),    # Water
        (145, 170, 100),  # Terrain
        (255, 255, 255),
    ]
    
    # Flatten the palette list and create a palette image
    palette_flattened = sum(palette, ())  # Flattens the list of tuples
    palette_img = Image.new('P', (1, 1))
    palette_img.putpalette(palette_flattened)
    
    with Image.open(image_path) as img:
        img = ImageEnhance.Brightness(img).enhance(4)  
        img = ImageEnhance.Contrast(img).enhance(4)    
        img = img.convert('P', palette=palette_img.palette)
        display(img)  

enhance_segmentation_image('/home/nfs/inf6/data/datasets/Carla_Moritz/SyncAngel3/Town03/seq_0003/segmentation_000.png')
```
